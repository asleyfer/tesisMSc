```python
X=df_manzanas.loc[:,['LATITUD','LONGITUD']]
X.head(100)
```

```python
kmeans = KMeans(n_clusters = 200).fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
centers = kmeans.cluster_centers_
```

```python
Labels=pd.DataFrame(y_kmeans, columns=list('z'))
```

```python
Lat_c_cluster = []
Lon_c_cluster = []
```

```python
Lat_c_cluster
```

```python
for index, row in df_manzanas.iterrows():
    s_value = row['Cluster'].astype(int)
    b_value = centers[s_value]
    Lat_c_cluster.append(b_value[0])
    Lon_c_cluster.append(b_value[1])
    
```

```python
df_manzanas['Lat_c_cluster'] = Lat_c_cluster
df_manzanas['Lon_c_cluster'] = Lon_c_cluster
```

```python
def haversine(row):
    lon1 = row['LONGITUD']
    lat1 = row['LATITUD']
    lon2 = row['Lon_c_cluster']
    lat2 = row['Lat_c_cluster']
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a)) 
    km = 6367 * c
    return km
```

```python
df_manzanas['distancia_a_clusteres'] = df_manzanas.apply(lambda row: haversine(row), axis=1)
```

```python
df_manzanas.to_csv("Manzanas_clusteres.csv",index=True)
```
